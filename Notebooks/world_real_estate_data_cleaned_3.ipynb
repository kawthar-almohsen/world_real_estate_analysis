```python

import re  # <-- this is the missing import

def extract_best_location(row):
    url = row['url']
    title = row['title']
    country = row['country']

    # Try from URL
    if pd.notna(url):
        parts = url.split('/')
        location_parts = [p.replace('-', ' ').title() for p in parts[5:-1] if p]
        if len(location_parts) >= 2:
            return ', '.join(location_parts)

    # Try from Title
    if pd.notna(title):
        match = re.search(r'in (.+)', title, re.IGNORECASE)
        if match:
            return match.group(1).strip()

    # Fallback to Country
    if pd.notna(country):
        return f"Unknown, {country}"

    return None

# Apply to missing locations only
missing_location = realestate_df['location'].isna()
realestate_df.loc[missing_location, 'location'] = realestate_df[missing_location].apply(extract_best_location, axis=1)
```

```python
def fill_by_group_median(df, group_col, target_col):
    # Step 1: Fill missing with group medians only if group has valid values
    group_medians = df.groupby(group_col)[target_col].median()

    def safe_fill(row):
        if pd.isna(row[target_col]):
            median = group_medians.get(row[group_col])
            return median if pd.notna(median) else row[target_col]
        return row[target_col]

    return df.apply(safe_fill, axis=1)
```

```python
# Bedrooms
realestate_df['apartment_bedrooms'] = fill_by_group_median(realestate_df, 'location', 'apartment_bedrooms')
realestate_df['apartment_bedrooms'].fillna(realestate_df['apartment_bedrooms'].median())
```

```python
def fill_year_by_group(df, group_cols, target_col):
    group_medians = df.groupby(group_cols)[target_col].median()

    def safe_fill(row):
        if pd.isna(row[target_col]):
            group_key = tuple(row[col] for col in group_cols) if isinstance(group_cols, list) else row[group_cols]
            median = group_medians.get(group_key)
            return median if pd.notna(median) else row[target_col]
        return row[target_col]

    return df.apply(safe_fill, axis=1)
```
